```python
from models import ConvolutionalNeuralNetworks
from utils import get_mnist_dataset, get_device, train, vis_losses_accs, visualize_predictions, plot_confusion_matrix
from torch.optim import SGD
import torch.nn as nn
```

```python
BATCH_SIZE = 32
train_dataset, train_data_loader = get_mnist_dataset(BATCH_SIZE)
```

```python
LR = 0.1
DEVICE = get_device()
model = ConvolutionalNeuralNetworks().to(DEVICE)
loss_function = nn.CrossEntropyLoss()
optimizer = SGD(model.parameters(), lr=LR)
```

```python
EPOCHS = 5
N_TRAIN_SAMPLES = len(train_dataset)
losses, accs = train(EPOCHS, train_data_loader, DEVICE, model, loss_function, optimizer, N_TRAIN_SAMPLES)
```

```python
vis_losses_accs(losses, accs)
```

```python
visualize_predictions(model, DEVICE, num_images=225)
```

```python
plot_confusion_matrix(model, DEVICE)
```
